# Fraud Detection - Exploratory Data Analysis (EDA)
This notebook performs Exploratory Data Analysis on the synthetic bank transaction dataset to understand fraud distributions and patterns.

```
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv('../data/transactions.csv')
print('Dataset Shape:', df.shape)
df.head()
```

## Class Balance
Let's look at the distribution of genuine vs fraudulent transactions.

```
class_counts = df['is_fraud'].value_counts()
class_pct = df['is_fraud'].value_counts(normalize=True) * 100
for cls, count, pct in zip(class_counts.index, class_counts, class_pct):
    label = 'Fraudulent' if cls == 1 else 'Genuine'
    print(f'{label}: {count} ({pct:.2f}%)')
```

## Feature Analysis by Fraud Class
Compare mean feature values for genuine vs fraudulent transactions to identify patterns.

```
df.groupby('is_fraud').mean()
```
